```python
import pandas as pd

df1 = pd.read_csv("processor_details.csv")
df2 = pd.read_csv("motherboard_details2.csv")
df3 = pd.read_csv("ssd_details2.csv")
df4 = pd.read_csv("gpu_details2.csv")
df5 = pd.read_csv("ram_details2.csv")

combined = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
combined.to_csv("product_details.csv", index=False)
```

```python
product_df = pd.read_csv("product_details.csv")
product_df.isnull().sum()
```

```python
df = product_df.dropna(subset=["Current Price", "Old Price"])
print(df.isnull().sum())
```

```python
price_cols = ["Current Price", "Old Price", "Discount Amount"]
for col in price_cols:
    df[col] = df[col].astype(str).str.replace("৳", "", regex=False).str.replace(",", "", regex=False).str.strip()
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df = df[~df["Product Name"].str.match(r'^\d[\d,\.]*৳?$', na=False)]

price_cols = ["Current Price", "Old Price", "Discount Amount"]
for col in price_cols:
    df = df[pd.to_numeric(df[col], errors="coerce").notna()]

df = df.reset_index(drop=True)
```

```python
df.to_csv("pc_products.csv", index=False)
```
